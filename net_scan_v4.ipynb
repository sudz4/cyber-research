```python
# libs
import nmap
import pandas as pd
```

Network segment #1
172.16.10.0/24

```python
def scan_network(subnet):
    nm = nmap.PortScanner()
    nm.scan(hosts=subnet, arguments='-sn -Pn')
    host_list = [item for item in nm.all_hosts() if nm[item].state() == 'up']
    return host_list

subnet = '172.16.10.0/24'
live_hosts = scan_network(subnet)
for host in live_hosts:
    print(f'Host {host} is up')
```

```python
def scan_network(subnet):
    nm = nmap.PortScanner()
    nm.scan(hosts=subnet, arguments='-sn -Pn')
    host_list = [item for item in nm.all_hosts() if nm[item].state() == 'up']
    return host_list

# List of subnets to scan
subnets = ['172.16.10.0/24', '172.16.1.0/24']

# Create an empty DataFrame
df = pd.DataFrame()

# Scan each subnet and store the results in the DataFrame
for subnet in subnets:
    live_hosts = scan_network(subnet)
    df_temp = pd.DataFrame(live_hosts, columns=[subnet])
    df = pd.concat([df, df_temp], axis=1)

# print(df)
df
```
